## Plotting Titanic Dataset

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
```

```python
titanic = pd.read_excel("../data/titanic.xls", "titanic")
titanic.head()
```

#### 1. Create a bar chart for number of passengers survived by class

```python
titanic.groupby('pclass').survived.sum().plot(kind='bar')
```

#### 2. Create a vertical bar chart for passengers survived by class and gender

```python
titanic.groupby(['sex','pclass']).survived.sum().plot(kind='barh')
```

#### 3. Create a stacked chart survived vs dead for different classes and gender

```python
deaths = pd.crosstab([titanic.pclass, titanic.sex], titanic.survived.astype(bool))
deaths.plot(kind='bar', stacked=True, color=['lightblue','lightgreen'], grid=False)
```

#### 4. Plot histogram of fares paid for passengers bins = 30

```python
titanic.fare.hist(bins=30, grid=False)
```

#### 5.Create a KDE vs bar plot of the fare

```python
titanic.fare.hist(bins=30, density=True, color='lightgreen')
titanic.fare.dropna().plot(kind='kde', xlim=(0,600), style='r--')
```

#### 6. Create a box plot of the fares by passenger class

```python
bpfares = titanic.boxplot(column= 'fare', by='pclass', grid=False)
for i in [1,2,3]:
    y = titanic.age[titanic.pclass==i].dropna()
    x = np.random.normal(i, 0.04, size=len(y))
    plt.plot(x, y, 'r.', alpha=0.2)
```

#### 7. Create a box plot of the age by passenger class

```python
bpage = titanic.boxplot(column='age', by='pclass', grid=False)
for i in [1,2,3]:
    y = titanic.age[titanic.pclass==i].dropna()
    x = np.random.normal(i, 0.04, size=len(y))
    plt.plot(x, y, 'r.', alpha=0.2)
```
