We are going to study the survival rate of passengers on titanic and what variables affected survival.

```python
from IPython.core.display import HTML
import os
import pandas as pd
HTML(filename='../data/titanic.html')
```

```python
# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=None)
t_df.rename(columns={0: 'pclass', 1: 'survived', 2: 'name', 3: 'sex', 4: 'age', 5: 'sibsp', 6: 'parch', 
                     7: 'ticket', 8: 'fare', 9: 'cabin', 10: 'embarked', 11: 'boat', 12: 'body', 13: 'home.dest'}, inplace=True)
t_df = t_df.drop(0)
t_df.head()
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (25 pts)*** 

```python
# group by survived
grouped = t_df.groupby('survived')
```

```python
# group by sex
grouped.get_group(1).groupby('sex')

#get total of survived
grouped.get_group(0).survived.count()

# group of amount of each gender who survived
total_group = grouped.get_group(1).groupby('sex')

# females_survived / total_survived = ratio
print(total_group.get_group('female').survived.count() / grouped.get_group(1).survived.count())

# males_survived / total_survived = ratio
print(total_group.get_group('male').survived.count() / grouped.get_group(1).survived.count())

# female survival ratio: 0.678
# male survival ratio: 0.322
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (40 pts)***

```python
# dividing the differnet age groups and naming them
child = t_df.loc[t_df['age'] < 14]
teen = t_df.loc[t_df['age'] >= 14]
teen = teen.loc[teen['age'] < 21]
adult = t_df.loc[t_df['age'] >= 21]
adult = adult.loc[adult['age'] < 65]
senior = t_df.loc[t_df['age'] >= 65]
```

```python
# create age range column
t_df["age_range"] = ""

# loop through to fill out new column properly
for x in range(1, len(t_df)+1):
    if (t_df.age[x] < 14.0):
        t_df.age_range[x] = "child"
    elif (t_df.age[x] >= 14.0) and (t_df.age[x] < 21.0):
        t_df.age_range[x] = "adolescent"
    elif (t_df.age[x] >= 21.0) and (t_df.age[x] < 65):
        t_df.age_range[x] = "adult"
    elif (t_df.age[x] >= 65.0):
        t_df.age_range[x] = "senior"
    else:
        t_df.age_range[x] = "NaN"
```
